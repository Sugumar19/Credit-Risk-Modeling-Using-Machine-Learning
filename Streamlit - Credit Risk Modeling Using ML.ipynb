```python
# 1️⃣ Importing Libraries
import streamlit as st # Used to create the web app interface.
import pandas as pd # Used to create dataframe for model input.
import joblib # Used to load saved model and encoders.

# 2️⃣ Load the Trained Model (This loads your trained XGBoost model from disk.) Without this: The app cannot predict & Model does not exist in memory
model = joblib.load("xgb_credit_model.pkl") # load the best model which is xgboost here

# 3️⃣ Load Saved Encoders: we must convert text → numbers exactly the same way
encoders = {col : joblib.load(f"{col}_encoder.pkl") for col in ["Sex", "Housing", "Saving accounts", "Checking account"]}
# loading model and encoders.  

# 4️⃣ App Title & Description --> This creates: Page title & Instruction text
st.title("Credit Risk Prediction App")
st.write("Enter applicant information to predict if the credit risk is good or bad")

# 5️⃣ Creating Input Fields: This collects user input
# Next we are going to define the input files
age = st.number_input("Age", min_value = 18, max_value = 80, value = 30) # default value is 30
# This ensures the age should 18 to 80. If no age mentioned, then default should be 30
sex = st.selectbox("Sex", ["male", "female"]) # Dropdown menu with 2 options: male & female
job = st.number_input("Job (0-3)", min_value = 0, max_value = 3, value = 1) # default value = 1
# This we have already encoded in dataset (0–3 categories).
housing = st.selectbox("Housing", ["own", "rent", "free"]) # Dropdown: own / rent / free
saving_accounts = st.selectbox("Saving accounts", ["little", "moderate", "rich", "quite rich"])
checking_account = st.selectbox("Checking account", ["little", "moderate", "rich"])
credit_amount = st.number_input("Credit Amount", min_value = 0, value = 12) # lets not put max value to this: no max limit
duration = st.number_input("Duration (months)", min_value = 1, value = 12) # To load Loan duration in months.

# 6️⃣ Prepare Input for Model --> input_df = pd.DataFrame({...})
# Now, lets prepare the input for the model
input_df = pd.DataFrame({
    "Age" : [age],
    "Sex" : [encoders["Sex"].transform([sex])[0]], # 
    "Job" : [job],
    7️⃣ Encoding Categorical Inputs
    "Housing" : [encoders["Housing"].transform([housing])[0]],
# encoders["Sex"].transform([sex])[0] --> 
    "Saving accounts" : [encoders["Saving accounts"].transform([saving_accounts])[0]],
    "Checking account" : [encoders["Checking account"].transform([checking_account])[0]],
    "Credit amount" : [credit_amount],
    "Duration" : [duration]
})

# 8️⃣ Prediction Button
# Next going to create a prediction button
if st.button("Predict Risk"):
    # 9️⃣ Make Prediction
    pred = model.predict(input_df)[0] # Model predicts: 0 or 1 ([0] extracts the value.)
    
    # 🔟 Show Result
    if pred == 1:
        st.success("The predicted credit risk is: **GOOD**")
    else:
        st.error("The predicted credit risk is: **BAD**")
# Streamlit shows colored message: Green for GOOD & Red for BAD
```
